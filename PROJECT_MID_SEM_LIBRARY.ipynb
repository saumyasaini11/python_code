```python
# Initialize library as a list of dictionaries
from datetime import date
library = []

# Function to add a book to the library
def add_book():
    book_id =int(input("Enter book ID: "))
    title = input("Enter the title of the book: ")
    author = input("Enter the author of the book: ")
    genre = input("Enter the genre of the book: ")
    issued_to=input("Enter issued to or enter None to skip: ")
    book = {id: book_id, "title": title, "author": author, "genre": genre, "issued_to": issued_to}
    library.append(book)
    print("Book added successfully!")
    print("\n")

# Function to update book details
def update_book():
    book_id = int(input("Enter the ID of the book to update: "))
    for book in library:
        if book[id] == book_id:
            title = input("Enter the updated title of the book: ")
            author = input("Enter the updated author of the book: ")
            genre = input("Enter the updated genre of the book: ")
            book["title"] = title
            book["author"] = author
            book["genre"] = genre
            print("Book details updated successfully!")
        else:
          print("Book not found!")

#Function to issue a book
def issue_book():
  book_id=int(input("Enter book ID to issue: "))
  for book in library:
    if book[id] == book_id and book["issued_to"] == "None":
      print("Book is available to issue. Please enter your details to proceed:")
      issued_to = input("Enter your name: ")
      update = {"issued_to": issued_to}
      book.update(update)
      sap_id = int(input("Enter your sap id: "))
      issued_on=date.today()
      print("Book issued on,",issued_on)
      print("Book issued successfully!")
    else:
      print('Book already issued or book not found!')
    print("\n")

# Function to re-issue a book
def reissue_book():
    book_id = int(input("Enter the ID of the book to re-issue: "))
    for book in library:
        if book[id] == book_id:
          print("Book is issued to: ", book["issued_to"])
          print("Book re-issued successfully!")
        else:
          print("Book not found!")

# Function to view book details
def view_books():
    for book in library:
        print("ID: ", book[id])
        print("Title: ", book["title"])
        print("Author: ", book["author"])
        print("Genre: ", book["genre"])
        if book["issued_to"] != "None":
            print("Issued to: ", book["issued_to"])
        print("\n")

# Function to remove a book
def remove_book():
    book_id = int(input("Enter the ID of the book to remove: "))
    for book in library:
        if book[id] == book_id:
            library.remove(book)
            print("Book removed successfully!")
        else:
          print("Book not found!")

# Main menu function
def main_menu():
    print("Library Management System")
    print("1. Add a book")
    print("2. Update book details")
    print("3. Issue a book")
    print("4. Re-issue a book")
    print("5. View books")
    print("6. Remove a book")
    print("7. Exit")

# Main program loop
while True:
    main_menu()
    choice = int(input("Enter your choice: "))
    if choice == 1:
        add_book()
    elif choice == 2:
        update_book()
    elif choice == 3:
        issue_book()
    elif choice == 4:
        reissue_book()
    elif choice == 5:
        view_books()
    elif choice == 6:
      remove_book()
    elif choice==7:
        break
```
